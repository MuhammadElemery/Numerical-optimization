## LR Full Implementation

### Step1: Initialize parameters (theta_0 & theta_1) with random value or simply zero. Also choose the Learning rate. 

```python
theta_0=0
theta_1=0
alpha=.001
```

### Step2: Use (theta_0 & theta_1) to predict the output h(x)= theta_0 + theta_1 * x.
#### Note: you will need to iterate through all data points

```python
m=len(x)
h=np.ones(m)*theta_0+theta_1*x
print(h)
```

### Step3: Calculate Cost function 𝑱(theta_0,theta_1 ).


```python
m=len(x)
print(m)
cost=np.sum((h-y)**2)*(1/2*m)
print(cost)
```

### Step4: Calculate the gradient.


```python
grad_0=1/m*np.sum(h-y)
grad_1=1/m * np.sum((h-y)*x)
print(grad_0)
print(grad_1)
```

### Step5: Update the parameters (simultaneously).


```python
theta0_new= theta_0-alpha*grad_0
theta1_one=theta_1-alpha*grad_1
```

### Step6: Repeat from 2 to 5 until converge to the minimum or achieve maximum iterations.

```python
max_iter=25
n=0
alpha=.005
while n <1000:
    n+=1
    
    h=theta_0+theta_1*x
    cost=np.sum((h-y)**2)*(1/2*m)
    
    grad_0=1/m*(np.sum(h-y))
    grad_1=1/m *( np.sum((h-y)*x))
    
    theta_0= theta_0-alpha*grad_0
    theta_1=theta_1-alpha*grad_1
    
    print(theta_0,theta_1,cost,n)

    
    if ((grad_0**2+grad_1**2)**(1/2) <0.0001):
        
        break
        
    


    

    
```

## Multivariate LR

#### Read MultipleLR csv file into numpy array  (check your data)
##### Data source
https://college.cengage.com/mathematics/brase/understandable_statistics/7e/students/datasets/slr/frames/frame.html


```python
data=np.genfromtxt("MultipleLR.csv",delimiter=",")
print(data)
```

```python
y=data[:,3]
m=len(y)
z=data[:,0:3]

```

```python
x0=np.ones(m)
print(x0)
```

```python
data_2=np.insert(z,0,x0,axis=1)
print(data_2)
```

```python
thetam=np.zeros(4)
```

### Repeat your implementation but for more than one variable

```python
m=len(y)
alpha=.0001
n=0
thetam=np.zeros(4)

while n<10000:
    n+=1
    hm= np.matmul(data_2, thetam)
    cost=hm-y

    gradm=1/m*(np.matmul(data_2.transpose(),cost))
    thetam=thetam-(alpha*gradm)
    
    if (np.linalg.norm(thetam))<.0001:
        break
    
    print(thetam,n,np.sum(cost))
    
    
```

#### Predict y values using the LR equation
##### h(x)= theta_0 + theta_1 * x1 + theta_2 * x2 + theta_3 * x3

```python
hm=np.matmul(thetam,data_2.transpose())
print(len(hm))
print(len(y))
```

### Plot loss vs. iterations

```python
m=len(y)
alpha=.00001
n=0
thetam=np.zeros(4)
iterations=[]
cost_lst=[]



while n<10000:
    n+=1
    hm= np.matmul(data_2, thetam)
    cost=hm-y

    gradm=1/m*(np.matmul(data_2.transpose(),cost))
    thetam=thetam-(alpha*gradm)
    
    iterations.append(n)
    cost_lst.append(np.sum(cost))
    
    
    if (np.linalg.norm(thetam))<.0001:
        break
        
plt.scatter(iterations,cost_lst)
plt.show()
    
    
    
    
```

#### Use R2 score to evaluate LR equation output

```python
r2_score(y,hm)
```
