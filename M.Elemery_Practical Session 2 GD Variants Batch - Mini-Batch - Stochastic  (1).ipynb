#### Use your code from practical session 1 and create a function to perform GD for single variable linear regression.
#### The function should take the following input parameters:
##### Input data (X), Target Labels (y), Learning Rate (alpha), Maximum number of iterations
#### The function should return the following:
##### All Theta_0 and Theta_1 values till converge, and equivellant Losses, all hypothesis output, and optimum values of Theta_0 and Theta_1
#### Note: you can choose any suitable stop criteria.

```python
def GD(x,y,alpha=.005,max_iter=10000):
    
    theta_0=0
    theta_1=0
    n=0
    cost_lst=[]
    iterations=[]
    m=len(x)
    theta_0lst=[]
    theta_1lst=[]
    
    while n < max_iter:
        n+=1
    
        h=theta_0+theta_1*x
        cost=(1/2*m)*(np.sum((h-y)**2))

        grad_0=(1/m)*(np.sum(h-y))
        grad_1=(1/m)*(np.sum((h-y)*x))

        theta_0=theta_0-(alpha*grad_0)
        theta_1=theta_1-(alpha*grad_1)
        
        cost_lst.append(cost)
        iterations.append(n)
        theta_0lst.append(theta_0)
        theta_1lst.append(theta_1)
        
        if ((grad_0**2+grad_1**2)**(1/2) <0.001):
            break
        
    return (cost_lst,iterations,theta_0lst,theta_1lst,h,n,theta_0,theta_1)
```

#### Update the implementation of the Batch GD to be Mini-Batch and rpeat the above plots and evaluations
#### Change batch size and see the effect

```python
def mini_batch(x,y,alpha=.005,max_iter=100,b=5):
    
    theta_0=0
    theta_1=0
    epoch=0
    n=0
    cost_lst=[]
    epochs=[]
    iterations=[]
    m=len(x)
    theta_0lst=[]
    theta_1lst=[]
    x.reshape(b,int(m/b))
    y.reshape(b,int(m/b))
    
    while epoch < max_iter:
        epoch+=1
        epochs.append(epoch)
        for i in range (b):
            n+=1
            
            h=theta_0+theta_1*x[i]
            cost=(1/2*(m/b))*(np.sum((h-y[i])**2))

            grad_0=(1/(m/b))*(np.sum(h-y[i]))
            grad_1=(1/(m/b))*(np.sum((h-y[i])*x[i]))

            theta_0=theta_0-(alpha*grad_0)
            theta_1=theta_1-(alpha*grad_1)
            
            iterations.append(n)

            cost_lst.append(cost)
            theta_0lst.append(theta_0)
            theta_1lst.append(theta_1)

        if ((grad_0**2+grad_1**2)**(1/2) <0.001):
            break

    return (cost_lst,epochs,iterations,theta_0lst,theta_1lst,h,n,theta_0,theta_1)
```

## Stochastic GD

#### Update the implementation of the Batch GD to beStochastic and rpeat the above plots and evaluations
#### You can try different learning rate and see its effect.
#### Compare to Batch and Mini-Batch GD

```python
def stochastic(x,y,alpha=.005,max_iter=1000):
    
    theta_0=0
    theta_1=0
    epoch=0
    n=0
    cost_lst=[]
    epochs=[]
    iterations=[]
    m=len(x)
    theta_0lst=[]
    theta_1lst=[]
    x.reshape(m,1)
    y.reshape(m,1)
    
    while epoch < max_iter:
        epoch+=1
        epochs.append(epoch)
        for i in range (m):
            n+=1
            
            h=theta_0+theta_1*x[i]
            cost=(1/2)*((h-y[i])**2)

            grad_0=(h-y[i])
            grad_1=(h-y[i])*x[i]

            theta_0=theta_0-(alpha*grad_0)
            theta_1=theta_1-(alpha*grad_1)
            
            iterations.append(n)

            cost_lst.append(cost)
            theta_0lst.append(theta_0)
            theta_1lst.append(theta_1)

        if ((grad_0**2+grad_1**2)**(1/2) <0.001):
            break

    return (cost_lst,epochs,iterations,theta_0lst,theta_1lst,h,n,theta_0,theta_1)
```
